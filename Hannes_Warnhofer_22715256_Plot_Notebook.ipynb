```python
# Author:
# Hannes Warnhofer
# 22715256
# hannes.warnhofer@fau.de

import os
import glob
import matplotlib.pyplot as plt
import pickle
import numpy as np
plt.rc('font', size=12)

# NOTICE: This will only work reasonable in my own folder structure and is only presnted in case it is needed for better comprehensability.
```

```python
def plot_multiple_histories(folder_path, file_pattern,text_str, labels=None):
    i=0
    fig, ax = plt.subplots(1,2, figsize=(12, 4))
    for file_path in glob.glob(os.path.join(folder_path, f'*{file_pattern}*.pkl')):
        
        with open(file_path, 'rb') as file:
            history = pickle.load(file)
            print(file)
        #label = labels[i] if labels and i < len(labels) else file
        ax[0].plot(history['accuracy'], label=labels[i],lw=2)
        ax[1].plot(history['val_accuracy'],lw=2)
        i+=1

    
    ax[0].set_xlabel("Epochs")
    ax[0].set_ylabel("Training Accuracy")

    ax[1].set_xlabel("Epochs")
    ax[1].set_ylabel("Validation Accuracy")

    ax[0].set_ylim(0.5,0.9)
    ax[1].set_ylim(0.5,0.9)


    ax[0].legend(loc='lower center',ncol=2)

    ax[0].grid('both')
    ax[1].grid('both')

    text_box = {
        'boxstyle': 'round',
        'facecolor': 'lightgrey',
        'edgecolor': 'black',
        'linewidth': 2,
        'alpha': 0.5,
        'pad': 0.5
    }
    plt.text(
        0.95, 0.05, text_str,
        transform=plt.gca().transAxes,
        ha='right', va='bottom',
        fontsize=12,
        bbox=text_box
    )
    
    #plt.gca().spines['top'].set_visible(False)
    #plt.gca().spines['right'].set_visible(False)
    #plt.gca().tick_params(axis='both', which='both', direction='out', top=False, right=False)

    #plt.subplots_adjust(left=0.08, bottom=0.08, right=0.92, top=0.92, wspace=0.2, hspace=0.2)
    #plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3), fancybox=True, shadow=True, ncol=2)
    plt.show()
    fig.savefig('Test_02-28-1945_batchsize512_epochs50_ratecheck.png', bbox_inches='tight')
```

```python
labels=["Rate 0.15","Rate 0.16","Rate 0.17","Rate 0.18","Rate 0.19","Rate 0.20","Rate 0.21","Rate 0.23","Rate 0.24","Rate 0.25"]
plot_multiple_histories("Histories/", "512batchsize","Batch Size: 512",labels)


```

```python
def plot_datasize_histories(folder_path, file_pattern,text_str,labels=None):
    i=0
    fig, ax = plt.subplots(1,2, figsize=(12, 4))
    for file_path in glob.glob(os.path.join(folder_path, f'*{file_pattern}*.pkl')):
        
        with open(file_path, 'rb') as file:
            history = pickle.load(file)
            print(file)
        #label = labels[i] if labels and i < len(labels) else file
        ax[0].plot(history['accuracy'], label=labels[i],lw=2)
        ax[1].plot(history['val_accuracy'],lw=2)
        i+=1

    
    ax[0].set_xlabel("Epochs")
    ax[0].set_ylabel("Training Accuracy")

    ax[1].set_xlabel("Epochs")
    ax[1].set_ylabel("Validation Accuracy")

    ax[0].set_ylim(0.5,0.9)
    ax[1].set_ylim(0.5,0.9)


    ax[0].legend(loc='lower center',ncol=2)

    ax[0].grid('both')
    ax[1].grid('both')

    text_box = {
        'boxstyle': 'round',
        'facecolor': 'lightgrey',
        'edgecolor': 'black',
        'linewidth': 2,
        'alpha': 0.5,
        'pad': 0.5
    }
    plt.text(
        0.95, 0.05, text_str,
        transform=plt.gca().transAxes,
        ha='right', va='bottom',
        fontsize=12,
        bbox=text_box
    )
    
    #plt.gca().spines['top'].set_visible(False)
    #plt.gca().spines['right'].set_visible(False)
    #plt.gca().tick_params(axis='both', which='both', direction='out', top=False, right=False)

    #plt.subplots_adjust(left=0.08, bottom=0.08, right=0.92, top=0.92, wspace=0.2, hspace=0.2)
    #plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3), fancybox=True, shadow=True, ncol=2)
    plt.show()
    fig.savefig('Test_02-28-2015_batchsize512_epochs50_datasizecheck.png', bbox_inches='tight')
```

```python
#labels=["143,000 events","190,000 events","237,000 events","284,000 events"]
labels=["284,000 events","237,000 events","190,000 events","143,000 events"]
plot_datasize_histories("Histories/EpochAnalysis/", "_50epochs","Dropout Rate: 0.25\nBatch Size: 512",labels)
```
